```python
# Def function to create the chart
def make_graph(stock_symbol):
    # WebScrapping the data:
    stock_data = yf.download(stock_symbol, period="1y", interval="1d")
    # Developing the chart code:
    plt.figure(figsize=(5,3))
    plt.plot(stock_data['Close'], label=f'{stock_symbol} - Close', color='green')
    plt.title(f'{stock_symbol} Stock Prices', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Closing Price (USD)', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()
# Using the function for Tesla Stock Prices:
make_graph("TSLA")
```

```python
# Using the function for Game Stop Prices:
make_graph("GME")
```
